# Linear Regression: Gradient Descent

In this notebook, we'll explore gradient descent as a memory-efficient alternative to direct methods for solving large-scale least squares problems. We'll:
1. Demonstrate the memory wall that direct methods hit
2. Implement gradient descent with matrix-vector products
3. Analyze convergence behavior and geometric properties
4. Compare performance across problem sizes

## Setup

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import time
import psutil
import os

# For reproducibility
torch.manual_seed(42)

# Plotting settings
plt.rcParams.update({
    'figure.figsize': (12, 6),
    'axes.titlesize': 14,
    'axes.labelsize': 12
})

def get_memory_gb():
    """Get current memory usage in GB"""
    process = psutil.Process()
    return process.memory_info().rss / 1e9

def get_process_memory():
    """Get current process memory usage in bytes"""
    process = psutil.Process()
    return process.memory_info().rss
```

## The Memory Wall

Let's start with a real-world example: genomic prediction with 1000 patients and 100,000 genetic markers.

```python
n_samples = 1000
n_markers = 100_000
memory_needed = (n_markers * n_markers * 8) / (1024**3)  # in GB
print(f"Memory needed for X^TX: {memory_needed:.1f} GB")  # 80.0 GB
```

Just forming $X^\top X$ would exceed most workstations' memory! Let's see this in practice with increasingly large problems:

```python
def measure_memory_and_time(n, p):
    """Measure memory usage and time for direct method."""
    print(f"\nProblem size: n={n}, p={p}")

    # Force CPU tensors
    torch.set_default_tensor_type(torch.FloatTensor)

    # Calculate total memory needed
    x_memory = 8 * n * p / 1e9  # GB for X
    xtx_memory = 8 * p * p / 1e9  # GB for X^T X
    total_memory = x_memory + xtx_memory

    # Get available system memory
    available = psutil.virtual_memory().available / 1e9  # GB
    current = get_memory_gb()

    print(f"Memory requirements:")
    print(f"  X: {x_memory:.1f}GB")
    print(f"  X^T X: {xtx_memory:.1f}GB")
    print(f"  Total needed: {total_memory:.1f}GB")
    print(f"  Currently used: {current:.1f}GB")
    print(f"  Available: {available:.1f}GB")

    # Check if we have enough memory (with 16GB safety margin)
    if total_memory + current + 16 > available:
        print(f"STOPPING: Would exceed available memory")
        return {
            'success': False,
            'error': 'Memory limit exceeded',
            'memory_needed': total_memory,
            'time': 0,
            'peak_memory': 0
        }

    # Try direct method with careful allocation
    start_mem = get_memory_gb()
    start_time = time.time()
    try:
        # First allocate X
        print("Allocating X...")
        X = torch.empty(n, p)  # Allocate without initializing
        if get_memory_gb() - start_mem > x_memory * 1.2:  # 20% margin
            raise MemoryError("X allocation exceeded estimate")
        X.normal_()  # Now initialize with random values

        # Then try X^T X
        print("Computing X^T X...")
        XtX = torch.empty(p, p)  # Pre-allocate
        if get_memory_gb() - start_mem > total_memory * 1.2:  # 20% margin
            del X, XtX  # Clean up
            raise MemoryError("X^T X allocation exceeded estimate")
        XtX = X.T @ X  # Compute

        print("Computing X^T y...")
        y = torch.randn(n)
        Xty = X.T @ y

        print("Solving system...")
        w = torch.linalg.solve(XtX, Xty)
        success = True

    except MemoryError as e:
        print(f"Memory allocation failed: {e}")
        success = False
    except RuntimeError as e:
        if 'out of memory' in str(e):
            print(f"PyTorch out of memory: {e}")
        else:
            print(f"Computation failed: {e}")
        success = False
    except Exception as e:
        print(f"Unexpected error: {e}")
        success = False
    finally:
        # Clean up any tensors
        try:
            del X, XtX, Xty, w
        except:
            pass

        # Force garbage collection
        import gc
        gc.collect()
        torch.cuda.empty_cache()  # Just in case

    time_taken = time.time() - start_time
    peak_mem = get_memory_gb() - start_mem

    return {
        'success': success,
        'error': None if success else 'Computation failed',
        'time': time_taken,
        'peak_memory': peak_mem,
        'memory_needed': total_memory
    }

# Test increasing sizes
sizes = [(1000, 1000), (1000, 5000), (1000, 20000), (1000, 30000)]
results = []
for n, p in sizes:
    result = measure_memory_and_time(n, p)
    results.append(result)
    if not result['success']:
        print("\nStopping due to memory limit exceeded")
        break

# Plot results
plt.figure(figsize=(12, 4))

# Memory plot
plt.subplot(121)
memory_needed = [r['memory_needed'] for r in results]
plt.plot(range(len(results)), memory_needed, 'o-', label='Memory needed')
plt.axhline(y=48, color='r', linestyle='--', label='Memory limit (48GB)')
plt.xticks(range(len(results)), [f'n={n}\np={p}' for n,p in sizes[:len(results)]], rotation=45)
plt.ylabel('Memory (GB)')
plt.title('Memory Requirements')
plt.legend()
plt.grid(True)

# Time plot
plt.subplot(122)
times = [r['time'] for r in results]
success = [r['success'] for r in results]
plt.plot(range(len(results)), times, 'o-')
for i, (t, s) in enumerate(zip(times, success)):
    color = 'g' if s else 'r'
    plt.plot(i, t, 'o', color=color)
plt.xticks(range(len(results)), [f'n={n}\np={p}' for n,p in sizes[:len(results)]], rotation=45)
plt.ylabel('Time (seconds)')
plt.title('Computation Time')
plt.grid(True)

plt.tight_layout()
plt.show()
```

The pattern is clear:
1. Small problems (p=1,000): Fast and memory-efficient
2. Medium problems (p=5,000): Still manageable
3. Large problems (p=20,000): Memory usage becomes concerning
4. Very large problems (p=50,000): Out of memory!

## A Memory-Efficient Alternative

Instead of forming $X^\top X$, gradient descent uses matrix-vector products. Let's compare the memory usage:

```python
def compare_operations(n, p):
    """Compare memory usage of matrix-matrix vs matrix-vector operations."""
    print(f"\nProblem size: n={n}, p={p}")

    # Generate data
    X = torch.randn(n, p)
    w = torch.randn(p)

    # Measure matrix-matrix product
    start_mem = get_memory_gb()
    try:
        XtX = X.T @ X
        print(f"X^TX memory: {get_memory_gb() - start_mem:.1f}GB")
    except RuntimeError as e:
        print(f"X^TX failed: {e}")

    # Measure matrix-vector products
    start_mem = get_memory_gb()
    try:
        Xw = X @ w
        XtXw = X.T @ Xw
        print(f"Matrix-vector memory: {get_memory_gb() - start_mem:.1f}GB")
    except RuntimeError as e:
        print(f"Matrix-vector failed: {e}")

# Compare on large problem
n, p = 1000, 20000
compare_operations(n, p)
```

The matrix-vector approach uses much less memory! Let's implement gradient descent:

```python
def gradient_descent(X, y, n_steps=100, step_size=None):
    """Solve least squares with gradient descent."""
    n, p = X.shape

    # Initialize weights
    w = torch.zeros(p)

    # Choose step size based on largest eigenvalue
    if step_size is None:
        # Get maximum eigenvalue for stepsize
        lambda_max = torch.linalg.eigvalsh(X.T @ X)[-1]
        step_size = 1.8 / lambda_max

    # Compute optimal value for relative loss
    w_opt = torch.linalg.lstsq(X, y).solution
    f_opt = 0.5 * ((X @ w_opt - y)**2).sum()
    f_init = 0.5 * ((X @ w - y)**2).sum()

    # Track loss
    losses = []

    # Main iteration
    for step in range(n_steps):
        # Compute gradient in two steps to avoid forming X^T X
        Xw = X @ w
        grad = X.T @ (Xw - y)

        # Take step
        w = w - step_size * grad

        # Track relative loss
        f_curr = 0.5 * ((X @ w - y)**2).sum()
        rel_loss = (f_curr - f_opt)/(f_init - f_opt)
        losses.append(rel_loss.item())

    return w, losses

# Test on medium-sized problem
n, p = 1000, 5000
print(f"\nTesting gradient descent (n={n}, p={p})")

# Generate data
X = torch.randn(n, p)
w_true = torch.randn(p)
y = X @ w_true

# Solve with gradient descent
start_time = time.time()
w_gd, losses = gradient_descent(X, y)
end_time = time.time()

print(f"Time: {end_time - start_time:.3f}s")

# Plot convergence
plt.figure(figsize=(10, 5))
plt.semilogy(losses)
plt.grid(True)
plt.xlabel('Iteration')
plt.ylabel('Relative Loss')
plt.title('Gradient Descent Convergence')
plt.show()
```

## Visualization in 2d

To understand why gradient descent works, let's visualize the optimization landscape. We'll start with a 2D problem where we can easily plot level sets:

```python
def plot_level_sets(X, y, w_path=None):
    """Plot level sets of least squares objective."""
    # Find optimal point to center plot
    w_opt = torch.linalg.lstsq(X, y).solution

    # Create grid of w values centered at optimal point
    grid_size = 2.0  # Size of grid around optimum
    w1, w2 = torch.meshgrid(
        torch.linspace(w_opt[0] - grid_size, w_opt[0] + grid_size, 100),
        torch.linspace(w_opt[1] - grid_size, w_opt[1] + grid_size, 100)
    )

    # Compute objective at each point
    f = torch.zeros_like(w1)
    for i in range(w1.shape[0]):
        for j in range(w1.shape[1]):
            w = torch.tensor([w1[i,j], w2[i,j]])
            f[i,j] = 0.5 * torch.norm(X @ w - y)**2

    # Plot level sets with more contours near minimum
    plt.figure(figsize=(10, 8))
    min_val = f.min()
    max_val = min_val + (f.max() - min_val) * 0.1  # Focus on region near minimum
    levels = torch.logspace(torch.log10(min_val), torch.log10(max_val), 30)
    plt.contour(w1, w2, f, levels=levels)
    plt.colorbar(label='Objective Value')

    # Plot optimization path if provided
    if w_path is not None:
        w_path = torch.stack(w_path)
        plt.plot(w_path[:,0], w_path[:,1], 'r.-', label='GD Path')
        plt.plot(w_path[0,0], w_path[0,1], 'g*', label='Start')
        plt.plot(w_path[-1,0], w_path[-1,1], 'r*', label='End')
        plt.plot(w_opt[0], w_opt[1], 'k*', label='Optimum', markersize=10)

        # Add arrows to show orthogonality
        for i in range(0, len(w_path)-1, 4):  # Show every 4th step
            w = w_path[i]
            # Compute gradient at this point
            grad = X.T @ (X @ w - y)
            grad = grad / torch.norm(grad)  # Normalize for visualization
            plt.arrow(w[0], w[1], -grad[0]*0.2, -grad[1]*0.2,
                     head_width=0.05, head_length=0.1, fc='r', ec='r', alpha=0.5)

    plt.xlabel('$w_1$')
    plt.ylabel('$w_2$')
    plt.title('Level Sets of Least Squares Objective')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

First, let's create a simple 2D problem and visualize it:

```python
# Generate 2D problem
n = 100
X = torch.randn(n, 2)
w_true = torch.tensor([1.0, -0.5])
y = X @ w_true + 0.1 * torch.randn(n)

# Run gradient descent and track path
def gradient_descent_with_path(X, y, n_steps=20):
    """Run gradient descent and return optimization path."""
    n, p = X.shape
    w = torch.zeros(p)
    path = [w.clone()]

    # Estimate step size
    v = torch.randn(p)
    v = v / torch.norm(v)
    for _ in range(10):
        v = X.T @ (X @ v)
        v = v / torch.norm(v)
    step_size = 1.8 / torch.norm(X @ v)**2

    for _ in range(n_steps):
        grad = X.T @ (X @ w - y)
        w = w - step_size * grad
        path.append(w.clone())

    return path

# Plot level sets with optimization path
path = gradient_descent_with_path(X, y)
plot_level_sets(X, y, path)
```

Now let's analyze how condition number and stepsize affect convergence:

```python
def create_test_problem(n=100, p=10, condition_number=10):
    # Create matrix with specified condition number
    U, _ = torch.linalg.qr(torch.randn(n, p))
    V, _ = torch.linalg.qr(torch.randn(p, p))
    s = torch.logspace(0, -torch.log10(torch.tensor(condition_number)), p)
    X = U @ torch.diag(s) @ V.T
    w_true = torch.randn(p)
    y = X @ w_true + 0.1 * torch.randn(n)
    return X, y, w_true

def plot_convergence_comparison():
    plt.figure(figsize=(15, 5))

    # Effect of step size (κ=10)
    plt.subplot(121)
    X, y, _ = create_test_problem(condition_number=10)
    lambda_max = torch.linalg.eigvalsh(X.T @ X)[-1]
    rel_steps = [0.1, 0.5, 1.8]  # relative to 1/lambda_max

    for rel_step in rel_steps:
        step_size = rel_step/lambda_max
        w = torch.ones(X.shape[1])  # Start from ones
        w_opt = torch.linalg.lstsq(X, y).solution
        f_opt = 0.5 * ((X @ w_opt - y)**2).sum()
        f_init = 0.5 * ((X @ w - y)**2).sum()

        losses = []
        for _ in range(100):
            Xw = X @ w
            grad = X.T @ (Xw - y)
            w = w - step_size * grad
            f_curr = 0.5 * ((X @ w - y)**2).sum()
            rel_loss = (f_curr - f_opt)/(f_init - f_opt)
            losses.append(rel_loss.item())
        plt.semilogy(losses, label=f'α = {rel_step}/λ_max')

    plt.grid(True)
    plt.xlabel('Iteration')
    plt.ylabel('Relative Loss (log scale)')
    plt.legend()
    plt.title('Effect of Step Size (κ=10)')

    # Effect of condition number
    plt.subplot(122)
    step_rel = 0.5  # use 0.5/lambda_max for all
    for kappa in [2, 10, 100]:
        X, y, _ = create_test_problem(condition_number=kappa)
        lambda_max = torch.linalg.eigvalsh(X.T @ X)[-1]

        w = torch.ones(X.shape[1])  # Start from ones
        w_opt = torch.linalg.lstsq(X, y).solution
        f_opt = 0.5 * ((X @ w_opt - y)**2).sum()
        f_init = 0.5 * ((X @ w - y)**2).sum()

        losses = []
        for _ in range(100):
            Xw = X @ w
            grad = X.T @ (Xw - y)
            w = w - step_rel/lambda_max * grad
            f_curr = 0.5 * ((X @ w - y)**2).sum()
            rel_loss = (f_curr - f_opt)/(f_init - f_opt)
            losses.append(rel_loss.item())
        plt.semilogy(losses, label=f'κ={kappa}')

    plt.grid(True)
    plt.xlabel('Iteration')
    plt.ylabel('Relative Loss (log scale)')
    plt.legend()
    plt.title('Effect of Condition Number\n(α = 0.5/λ_max)')

    plt.tight_layout()
    plt.show()

def plot_2d_zigzag():
    plt.figure(figsize=(12, 12))

    kappas = [2, 50]  # Small and large condition numbers
    alphas = [0.1, 1.8]  # Small and large relative stepsizes

    for i, kappa in enumerate(kappas):
        for j, alpha_rel in enumerate(alphas):
            plt.subplot(2, 2, 2*i + j + 1)

            # Create eigenvectors and eigenvalues explicitly
            v1 = torch.tensor([1., 1.], dtype=torch.float32)  # First eigenvector
            v1 = v1 / torch.norm(v1)
            v2 = torch.tensor([-1., 1.], dtype=torch.float32)  # Second eigenvector
            v2 = v2 / torch.norm(v2)

            # Form V (eigenvector matrix) and D (eigenvalue matrix)
            V = torch.stack([v1, v2], dim=1)  # Stack as columns
            lambda1 = torch.tensor(1.0, dtype=torch.float32)
            lambda2 = torch.tensor(1.0/kappa, dtype=torch.float32)
            D = torch.diag(torch.tensor([lambda1, lambda2]))

            # Form X^T X directly (we only need this for the quadratic form)
            XtX = V @ D @ V.T

            # Set optimum at origin by setting y = 0
            w_true = torch.zeros(2)
            y = torch.zeros(2)

            # Create contour plot
            w1, w2 = torch.meshgrid(
                torch.linspace(-2, 2, 50, dtype=torch.float32),
                torch.linspace(-2, 2, 50, dtype=torch.float32)
            )
            Z = torch.zeros_like(w1)
            for i1 in range(w1.shape[0]):
                for i2 in range(w1.shape[1]):
                    w = torch.tensor([w1[i1,i2], w2[i1,i2]], dtype=torch.float32)
                    Z[i1,i2] = 0.5 * w @ XtX @ w  # Quadratic form (y=0 so no linear term)

            plt.contour(w1, w2, Z, levels=20)

            # Start from a point that's a mixture of eigenvectors
            # Use v1 + v2 normalized to ensure equal components
            w_start = (v1 + v2)
            w_start = w_start * (2.0 / torch.norm(w_start))  # Scale to be at distance 2
            w = w_start
            path = [w.clone()]

            for _ in range(50):  # More iterations to show zigzag
                grad = XtX @ w  # Gradient for quadratic form (y=0)
                w = w - alpha_rel * grad  # Note: alpha already accounts for lambda_max=1
                path.append(w.clone())

            path = torch.stack(path)
            plt.plot(path[:,0], path[:,1], 'r.-', label='GD path')
            plt.plot(0, 0, 'g*', markersize=10, label='Optimum')

            plt.axis('equal')
            plt.title(f'κ = {kappa}, α = {alpha_rel}/λ_max')
            if i == 0 and j == 0:
                plt.legend()

            # Add eigenvector directions
            plt.arrow(0, 0, v1[0], v1[1], color='gray', alpha=0.3,
                     head_width=0.1, head_length=0.1)
            plt.arrow(0, 0, v2[0], v2[1], color='gray', alpha=0.3,
                     head_width=0.1, head_length=0.1)

    plt.tight_layout()
    plt.show()

# Run visualization
plot_2d_zigzag()
```

Key observations:
1. Well-conditioned (κ=2):
   - Small α: Slow but steady progress
   - Large α: Fast, direct path
2. Poorly-conditioned (κ=50):
   - Small α: Very slow progress
   - Large α: Zigzag path, many wasted steps

## Stepsize Selection and Convergence Analysis

The stepsize bound $0 < \alpha_k < \frac{2}{\lambda_{\max}(X^\top X)}$ is crucial for convergence. Let's see why:

```python
def analyze_stepsize(X, y, stepsizes, n_steps=100):
    """Compare convergence for different stepsizes."""
    # Get maximum eigenvalue for reference
    lambda_max = torch.linalg.eigvalsh(X.T @ X)[-1]

    # Compute optimal value for relative loss
    w_opt = torch.linalg.lstsq(X, y).solution
    f_opt = 0.5 * ((X @ w_opt - y)**2).sum()

    plt.figure(figsize=(12, 5))

    # Create initial point (same for all runs)
    w_init = torch.ones(X.shape[1])  # Start from all ones
    f_init = 0.5 * ((X @ w_init - y)**2).sum()

    for alpha_rel in stepsizes:
        # Run gradient descent
        w = w_init.clone()  # Start from same point
        losses = []

        step_size = alpha_rel / lambda_max

        for _ in range(n_steps):
            # Compute gradient in two steps
            Xw = X @ w
            grad = X.T @ (Xw - y)
            w = w - step_size * grad

            # Track relative loss
            f_curr = 0.5 * ((X @ w - y)**2).sum()
            rel_loss = (f_curr - f_opt)/(f_init - f_opt)
            losses.append(rel_loss)

        plt.semilogy(losses, label=f'α = {alpha_rel}/λ_max')

    plt.grid(True)
    plt.xlabel('Iteration')
    plt.ylabel('Relative Loss (log scale)')
    plt.title('Effect of Stepsize on Convergence')
    plt.legend()
    plt.show()

    return lambda_max

# Create well-conditioned problem
n, p = 100, 20
X = torch.randn(n, p)
w_true = torch.randn(p)
y = X @ w_true + 0.1 * torch.randn(n)

# Test different stepsizes
lambda_max = analyze_stepsize(X, y, [0.1, 0.5, 1.8])
print(f"\nMaximum eigenvalue: {lambda_max:.2f}")

def analyze_conditioning(n, p, kappas, n_steps=100):
    """Compare convergence for different condition numbers."""
    plt.figure(figsize=(12, 5))

    for kappa in kappas:
        # Create problem with specified condition number
        U, _ = torch.linalg.qr(torch.randn(n, p))
        V, _ = torch.linalg.qr(torch.randn(p, p))
        s = torch.logspace(0, -torch.log10(torch.tensor(kappa)), p)
        X = U @ torch.diag(s) @ V.T
        w_true = torch.randn(p)
        y = X @ w_true + 0.1 * torch.randn(n)

        # Get optimal value for relative loss
        w_opt = torch.linalg.lstsq(X, y).solution
        f_opt = 0.5 * ((X @ w_opt - y)**2).sum()

        # Run gradient descent
        w = torch.zeros(p)
        losses = []
        f_init = 0.5 * ((X @ w - y)**2).sum()

        # Use optimal step size
        lambda_max = torch.linalg.eigvalsh(X.T @ X)[-1]
        step_size = 0.5 / lambda_max  # Use 0.5/lambda_max for stability

        for _ in range(n_steps):
            Xw = X @ w
            grad = X.T @ (Xw - y)
            w = w - step_size * grad

            f_curr = 0.5 * ((X @ w - y)**2).sum()
            rel_loss = (f_curr - f_opt)/(f_init - f_opt)
            losses.append(rel_loss)

        plt.semilogy(losses, label=f'κ = {kappa}')

    plt.grid(True)
    plt.xlabel('Iteration')
    plt.ylabel('Relative Loss (log scale)')
    plt.title('Effect of Condition Number on Convergence')
    plt.legend()
    plt.show()

# Compare different condition numbers
n, p = 100, 20
analyze_conditioning(n, p, [2, 10, 100])
```

Key insights:
1. Well-conditioned (κ=2): Fast, steady convergence
2. Moderate conditioning (κ=10): Slower but stable
3. Poorly-conditioned (κ=100): Much slower convergence

## Limitations and Future Directions


Three key limitations emerge:

1. **Computation Cost**: Each iteration needs $O(np)$ operations
   - For large $n$: Computing full gradient expensive
   - Solution: Stochastic methods (sample subset each iteration)

2. **Memory Scaling**: Still needs $O(np)$ memory
   - For large $p$: Memory can still be an issue
   - Solution: Coordinate descent (update subset of parameters)

3. **Convergence Speed**: Depends heavily on condition number
   - For ill-conditioned problems: Very slow convergence
   - Solution: Momentum and adaptive methods

We'll explore these solutions in future lectures:
1. Stochastic gradient descent for large $n$
2. Coordinate descent for large $p$
3. Momentum and adaptive methods for poor conditioning
